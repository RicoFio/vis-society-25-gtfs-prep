# Stops

```python
stops = gpd.read_file('./original_data/mbta_stops.geojson')
stops = stops.drop(columns=['stop_desc', 'location_type', 'agency_name', 'platform_code'])
stops_with_buffer = gpd.read_file('./original_data/mbta_stops_with_buffer.geojson')
stops_with_buffer = stops_with_buffer.drop(columns=['stop_desc', 'location_type', 'agency_name', 'platform_code'])
```

```python
stops.to_csv('./mbta_stops/mbta_stops.geojson')
stops_with_buffer.to_csv('./mbta_stops_with_buffer/mbta_stops_with_buffer.geojson')
```

```python
dup_lines = lines[lines.route_long_name.isin(lines[lines.route_long_name.duplicated()].route_long_name)]
colors = [matplotlib.colors.rgb2hex(rgba) for rgba in cm.rainbow(np.linspace(0, 1, len(dup_lines)))]
dup_lines.explore(color=colors)
```

```python
dup_lines = lines[lines.route_long_name.duplicated()]
colors = [matplotlib.colors.rgb2hex(rgba) for rgba in cm.rainbow(np.linspace(0, 1, len(dup_lines)))]
dup_lines.explore(color=colors)
```

```python
lines_deduplicated = lines[~lines.route_long_name.duplicated()]
```
